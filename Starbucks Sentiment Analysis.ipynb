```python
#adding a sentiment column since i finally need one for nlp
clean['Sentiment'] = clean['Rating'].apply(lambda x:1 if x >= 3 else (0 if x < 3 else -1))
clean
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords.words('english')
```

```python
def mani(text):
    tl = [ch for ch in text if ch not in string.punctuation]
    ts = ''.join(tl)
    tlfinal = [word for word in ts.split() if word.lower() not in stopwords.words('english')]
    return tlfinal
```

```python
t = "hey! how is it going? life has been really, truly weird"
mani(t)
```

```python
c = clean.Review.apply(mani)
c
```

```python
t = ['hello', 'hello word', 'hello hello word word']
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X = v.fit_transform(t)
print(v.get_feature_names_out())
print(X.toarray())
```

```python
v1 = CountVectorizer(analyzer = mani)
ccv = v1.fit_transform(clean.Review)
ccv
```
